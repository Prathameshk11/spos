```python
import tkinter as tk
from tkinter import messagebox

# Function to handle sign-up submission
def submit_signup():
    # Get data from entries
    username = username_entry.get()
    password = password_entry.get()
    email = email_entry.get()

    # Validation (simple example)
    if not username or not password or not email:
        messagebox.showwarning("Warning", "All fields are required.")
        return

    # Show confirmation message (this is where you'd handle actual sign-up logic)
    messagebox.showinfo("Sign-Up Successful", f"Welcome, {username}!")
    
    # Clear the fields after submission
    username_entry.delete(0, tk.END)
    password_entry.delete(0, tk.END)
    email_entry.delete(0, tk.END)

# Initialize main window
root = tk.Tk()
root.title("Sign-Up Form")
root.geometry("400x300")
root.configure(bg="#e0f7fa")

# Form title
title_label = tk.Label(root, text="Sign-Up", font=("Helvetica", 16, "bold"), bg="#e0f7fa")
title_label.pack(pady=10)

# Username
username_label = tk.Label(root, text="Username:", bg="#e0f7fa")
username_label.pack()
username_entry = tk.Entry(root, width=30)
username_entry.pack(pady=5)

# Password
password_label = tk.Label(root, text="Password:", bg="#e0f7fa")
password_label.pack()
password_entry = tk.Entry(root, show="*", width=30)
password_entry.pack(pady=5)

# Email
email_label = tk.Label(root, text="Email:", bg="#e0f7fa")
email_label.pack()
email_entry = tk.Entry(root, width=30)
email_entry.pack(pady=5)

# Sign-Up Button
signup_button = tk.Button(root, text="Sign Up", command=submit_signup, bg="#00796b", fg="white", width=10)
signup_button.pack(pady=20)

# Start main loop
root.mainloop()
```
