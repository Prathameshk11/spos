```python
import tkinter as tk
from tkinter import ttk
from tkinter import messagebox

```

```python
# Function to submit the form
def submit_form():
    # Get data from entries
    name = name_entry.get()
    age = age_entry.get()
    gender = gender_var.get()
    email = email_entry.get()
    phone = phone_entry.get()
    sport = sport_var.get()

    # Validation (simple example, expand as needed)
    if not name or not age or not email or not phone or not sport:
        messagebox.showwarning("Warning", "Please fill all the required fields")
        return

    # Show confirmation message
    messagebox.showinfo("Submission Successful", f"Thank you {name} for registering!")

    # Clear the fields after submission
    name_entry.delete(0, tk.END)
    age_entry.delete(0, tk.END)
    email_entry.delete(0, tk.END)
    phone_entry.delete(0, tk.END)
    gender_var.set("Select Gender")
    sport_var.set("Select Sport")

# Initialize main window
root = tk.Tk()
root.title("Sports Academy Registration Form")
root.geometry("400x400")
root.configure(bg="#e6e6fa")

# Form title
title_label = tk.Label(root, text="Sports Academy Registration", font=("Helvetica", 16, "bold"), bg="#e6e6fa")
title_label.pack(pady=10)

# Name
name_label = tk.Label(root, text="Full Name:", bg="#e6e6fa")
name_label.pack()
name_entry = tk.Entry(root, width=30)
name_entry.pack(pady=5)

# Age
age_label = tk.Label(root, text="Age:", bg="#e6e6fa")
age_label.pack()
age_entry = tk.Entry(root, width=30)
age_entry.pack(pady=5)

# Gender
gender_var = tk.StringVar()
gender_var.set("Select Gender")
gender_label = tk.Label(root, text="Gender:", bg="#e6e6fa")
gender_label.pack()
gender_dropdown = ttk.Combobox(root, textvariable=gender_var, values=["Male", "Female", "Other"], state="readonly")
gender_dropdown.pack(pady=5)

# Email
email_label = tk.Label(root, text="Email:", bg="#e6e6fa")
email_label.pack()
email_entry = tk.Entry(root, width=30)
email_entry.pack(pady=5)

# Phone
phone_label = tk.Label(root, text="Phone:", bg="#e6e6fa")
phone_label.pack()
phone_entry = tk.Entry(root, width=30)
phone_entry.pack(pady=5)

# Sport Selection
sport_var = tk.StringVar()
sport_var.set("Select Sport")
sport_label = tk.Label(root, text="Preferred Sport:", bg="#e6e6fa")
sport_label.pack()
sport_dropdown = ttk.Combobox(root, textvariable=sport_var, values=["Football", "Basketball", "Tennis", "Swimming", "Athletics"], state="readonly")
sport_dropdown.pack(pady=5)

# Submit Button
submit_button = tk.Button(root, text="Submit", command=submit_form, bg="#4CAF50", fg="white")
submit_button.pack(pady=20)

# Start main loop
root.mainloop()
```
